```python
import pandas as pd
import numpy as np
import re
import os
```

## Spam

```python
iter = 0
subject = []
body = []
tobe_body = ''
while (True):
    try:
        with open("data/spam/mail"+str(iter)+".txt","rb") as f:
            for i,line in enumerate(f):
                line = line.rstrip()
                if i==0:
                    line = re.sub("b'Subject: ","",str(line))
                    subject.append(line)
                else:
                    line = re.sub("b'","",str(line))
                    line = re.sub("'b","",line)

                    tobe_body += line
                    
            body.append(tobe_body)
            tobe_body=''
            iter+=1
        
    except FileNotFoundError:
        print(str(iter) +" spam files")
        break
    
```

```python
assert (len(subject) == len(body)  ), "It's not  correct"
```

```python
df = pd.DataFrame(columns=['subject','body','label'])
```

```python
df.subject = subject
```

```python
df.body = body
```

```python
df.label = 1
```

```python
df.to_csv('csv/spam.csv')
```

## users- inbox

```python
iter = 0
usr_iter = 1
subject = []
body = []
tobe_body = ''
while (True):
    try:
        with open("data/user"+ str(usr_iter)+"/inbox/mail"+str(iter)+".txt","rb") as f:
            how_many_file = os.listdir("data/user"+ str(usr_iter)+"/inbox")
            number_files = len(how_many_file)
            for i,line in enumerate(f):
                line = line.rstrip()
                if i==0:
                    line = re.sub("b'Subject: ","",str(line))
                    subject.append(line)
                else:
                    line = re.sub("b'","",str(line))
                    line = re.sub("'b","",line)

                    tobe_body += line
                    
            body.append(tobe_body)
            tobe_body=''
            iter+=1
            
            if iter == number_files-1:
                usr_iter+=1
                iter=0
            
    except FileNotFoundError:
        break
```

```python
df = pd.DataFrame(columns=['subject','body','label'])

df.subject = subject
df.body = body
df.label=0
```

```python
df.to_csv('csv/not_spam.csv')
```
